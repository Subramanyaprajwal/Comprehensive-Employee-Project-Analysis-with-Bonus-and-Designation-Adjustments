```python
import pandas as pd

data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, '', 680000, 400000, 350000, '', 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}

Project = pd.DataFrame(data)

Project['Cost'] = pd.to_numeric(Project['Cost'], errors='coerce')

running_sum = 0
count = 0
for i in range(len(Project)):
    if pd.isnull(Project['Cost'][i]):
        Project.loc[i, 'Cost'] = running_sum / count if count > 0 else 0
    else:
        running_sum += Project['Cost'][i]
        count += 1

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Jhony'],
    'Gender': ['M','F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}

Employee = pd.DataFrame(data)

# Splitting 'Name' column into 'First Name' and 'Last Name' columns
Employee[['First Name', 'Last Name']] = Employee['Name'].str.split(' ', expand=True)

# Dropping the original 'Name' column
Employee.drop(columns=['Name'], inplace=True)

print(Employee)
```

```python
merged_df = pd.merge(Project, Employee, on='ID', how='inner')

# Merging the resulting dataframe with Seniority_level based on the 'ID' column
Final = pd.merge(merged_df, Seniority_level, on='ID', how='inner')

print(Final)
```

```python
Final['Cost'] = pd.to_numeric(Final['Cost'], errors='coerce')

# Adding a 'Bonus' column based on conditions
Final['Bonus'] = 0  # Initialize the 'Bonus' column with default value

# Applying conditions to calculate bonus
condition = (Final['Status'] == 'Finished')  # Considering only finished projects
Final.loc[condition, 'Bonus'] = Final.loc[condition, 'Cost'] * 0.05  # 5% bonus for finished projects

print(Final)
```

```python
condition_fail = Final['Status'] == 'Failed'
Final.loc[condition_fail, 'Designation level'] -= 1

condition_above_4 = Final['Designation level'] > 4
Final = Final[~condition_above_4] 

print(Final)
```

```python
Final.loc[Final['Gender'] == 'M', 'First Name'] = 'Mr. ' + Final['First Name']
Final.loc[Final['Gender'] == 'F', 'First Name'] = 'Mrs. ' + Final['First Name']

Final.drop(columns=['Gender'], inplace=True)

print(Final)
```

```python
condition_age = Final['Age'] > 29
Final.loc[condition_age, 'Designation level'] += 1

print(Final)
```

```python
TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)

print(TotalProjCost)
```
